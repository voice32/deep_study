```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
size = 5
```

# Linear activation

```python
inputs = linear = np.linspace(-10., 10., 1000)

plt.figure(figsize=(size,size))
plt.title('Linear activation', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, linear)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Sigmoid

```python
inputs = np.linspace(-10., 10., 1000)
sigmoid = 1 / (1 + np.exp(-inputs))

plt.figure(figsize=(size,size))
plt.title('Sigmoid', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, sigmoid)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Hard Sigmoid

```python
inputs = np.linspace(-10., 10., 1000)
hard_sigmoid = [0.] * 375 + [0.004 * a for a in range(250)] + [1.] * 375

plt.figure(figsize=(size,size))
plt.title('Hard Sigmoid', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, hard_sigmoid)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Hyperbolic Tangent (TanH)

```python
inputs = np.linspace(-10., 10., 1000)
tanh = (np.exp(inputs) - np.exp(-inputs)) / (np.exp(inputs) + np.exp(-inputs))

plt.figure(figsize=(size,size))
plt.title('Hyperbolic Tangent (TanH)', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, tanh)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# SoftSign

```python
inputs = np.linspace(-10., 10., 1000)
softsign = inputs / (1 + np.abs(inputs))

plt.figure(figsize=(size,size))
plt.title('SoftSign activation', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, softsign)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Rectified Linear Unit (ReLU)

```python
def relu(x):
    if x >= 0:
        return x
    else:
        return 0

inputs = np.linspace(-10., 10., 1000)
v_relu = [relu(x) for x in inputs]

plt.figure(figsize=(size,size))
plt.title('Rectified Linear Unit (ReLU)', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, v_relu)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Leaky ReLU

```python
def leaky_relu(x, alpha):
    if x >= 0:
        return x
    else:
        return x * alpha

inputs = np.linspace(-10., 10., 1000)
leaky_relu_001 = [leaky_relu(x, 0.01) for x in inputs]
leaky_relu_01 = [leaky_relu(x, 0.1) for x in inputs]
leaky_relu_03 = [leaky_relu(x, 0.3) for x in inputs]

plt.figure(figsize=(size,size))
plt.title('Leaky ReLU', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, leaky_relu_001, label='alpha = 0.01')
plt.plot(inputs, leaky_relu_01, label='alpha = 0.1')
plt.plot(inputs, leaky_relu_03, label='alpha = 0.3')

plt.legend(loc='upper left')

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Thresholded ReLU

```python
def thresholded_relu(x, theta):
    if x < theta:
        return 0
    else:
        return x

thr = 1.5
inputs_1 = np.linspace(-10., thr, 10)
values_1 = [0.]*10
inputs_2 = [thr, thr+0.01]
values_2 = [0, thr]
inputs_3 = values_3 = np.linspace(thr, 10., 10)

plt.figure(figsize=(size,size))
plt.title('Thresholded ReLU', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs_1, values_1, color = 'b')
plt.plot(inputs_2, values_2, linestyle = '--', color = 'b')
plt.plot(inputs_3, values_3, color = 'b')

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Exponential Linear Unit (ELU)

```python
def elu(x, alpha):
    if x > 0:
        return x
    else:
        return alpha * (np.exp(x) - 1)
    
inputs = np.linspace(-10., 10., 1000)
elu_1 = [elu(x, 1.) for x in inputs]
elu_05 = [elu(x, 0.5) for x in inputs]
elu_15 = [elu(x, 1.5) for x in inputs]

plt.figure(figsize=(size,size))
plt.title('Exponential Linear Unit (ELU)', fontsize=14)
plt.plot(inputs, elu_05, label = 'alpha = 0.5')
plt.plot(inputs, elu_1, label = 'alpha = 1.0')
plt.plot(inputs, elu_15, label = 'alpha = 1.5')
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')

plt.legend(loc='upper left')
plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Scaled Exponential Linear Unit (SELU)

```python
def selu(x, lmbd = 1.0507, alpha = 1.6733):
    if x > 0:
        return lmbd * x
    else:
        return lmbd * alpha * (np.exp(x) - 1)
    
inputs = np.linspace(-10., 10., 1000)
v_selu = [selu(x) for x in inputs]

plt.figure(figsize=(size,size))
plt.title('Scaled Exponential Linear Unit (SELU)', fontsize=14)
plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, v_selu)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# SoftPlus

```python
inputs = np.linspace(-10., 10., 1000)
softplus = [np.log(1+np.exp(x)) for x in inputs]

plt.figure(figsize=(size,size))
plt.title('SoftPlus', fontsize=14)

plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')
plt.plot(inputs, softplus)

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```

# Comparing activations

```python
inputs = linear = np.linspace(-5., 5., 1000)
sigmoid = 1 / (1 + np.exp(-inputs))
hard_sigmoid = [0.] * 375 + [0.004 * a for a in range(250)] + [1.] * 375
tanh = (np.exp(inputs) - np.exp(-inputs)) / (np.exp(inputs) + np.exp(-inputs))
softsign = inputs / (1 + np.abs(inputs))

thr = 1.5
thr_inputs_1 = np.linspace(-10., thr, 10)
thr_values_1 = [0.]*10
thr_inputs_2 = [thr, thr+0.01]
thr_values_2 = [0, thr]
thr_inputs_3 = thr_values_3 = np.linspace(thr, 10., 10)


plt.figure(figsize=(size * 3, size * 1.5))
plt.suptitle('Comparing activation functions', fontsize=14)
plt.subplot(1,2,1)
plt.title('Sigmoid-like activations')

plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')

plt.plot(inputs, sigmoid, label='Sigmoid')
plt.plot(inputs, hard_sigmoid, label='Hard Sigmoid')
plt.plot(inputs, tanh, label='TanH')
plt.plot(inputs, softsign, label='SoftSign')

plt.legend(loc='upper left')

plt.xlabel('Input')
plt.ylabel('Output')

inputs = linear = np.linspace(-5., 5., 1000)
v_relu = [relu(x) for x in inputs]
leaky_relu_03 = [leaky_relu(x, 0.3) for x in inputs]
thr = 1.5
thr_inputs_1 = np.linspace(-5., thr, 10)
thr_values_1 = [0.]*10
thr_inputs_2 = [thr, thr+0.01]
thr_values_2 = [0, thr]
thr_inputs_3 = thr_values_3 = np.linspace(thr, 5., 10)

elu_1 = [elu(x, 1.0) for x in inputs]
v_selu = [selu(x) for x in inputs]
softplus = [np.log(1+np.exp(x)) for x in inputs]

plt.subplot(1,2,2)
plt.title('Linear and Rectified activations')

plt.axvline(x=0, color = '#CCCCCC', linestyle='--')
plt.axhline(y=0, color = '#CCCCCC', linestyle='--')

plt.plot(inputs, linear, label='Linear')
plt.plot(inputs, v_relu, label='ReLU')
plt.plot(inputs, leaky_relu_03, label='Leaky ReLU, alpha = 0.3')

plt.plot(thr_inputs_1, thr_values_1, color = 'b', label='Thresholded ReLU, theta = 1.5')
plt.plot(thr_inputs_2, thr_values_2, linestyle = '--', color = 'b')
plt.plot(thr_inputs_3, thr_values_3, color = 'b')

plt.plot(inputs, elu_1, label = 'ELU, alpha = 1.0')
plt.plot(inputs, v_selu, label='SELU')
plt.plot(inputs, softplus, label='SoftPlus')

plt.legend(loc='upper left')

plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```
