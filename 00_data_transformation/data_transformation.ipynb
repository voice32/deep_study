```python
import keras
import keras.utils as kutils
from keras.datasets import mnist
import numpy as np
from scipy import stats
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
plt.subplot(221)
plt.imshow(x_train[0], cmap=plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(x_train[1], cmap=plt.get_cmap('gray'))
plt.subplot(223)
plt.imshow(x_train[2], cmap=plt.get_cmap('gray'))
plt.subplot(224)
plt.imshow(x_train[3], cmap=plt.get_cmap('gray'))

plt.show()
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
for i in range(len(x_train)):
    if max(x_train[i]) < 255:
        print('Found image with index %d with no 0 and 255 values' % i)
        print(stats.describe(x_train[i]))
        

stats.describe(x_train.reshape(60000 * 784, ))
```

```python
plt.subplot(221)
plt.imshow(x_train[10], cmap=plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(x_train[24], cmap=plt.get_cmap('gray'))
plt.subplot(223)
plt.imshow(x_train[46], cmap=plt.get_cmap('gray'))
plt.subplot(224)
plt.imshow(x_train[58], cmap=plt.get_cmap('gray'))

plt.show()
```

```python
n_train = x_train / 255.
for i in range(4):
    print(stats.describe(n_train[i]))

n_train = n_train.reshape(60000 * 784, )
stats.describe(n_train)
```

```python
def normalize_sample_wise(dataset):
    normalized = []
    for s in dataset:
        mn, mx = s.min(), s.max()
        s = (s - mn) / (mx - mn)
        normalized.append(s)
    return np.array(normalized)

nsw_train = normalize_sample_wise(x_train)

for i in range(4):
    print(stats.describe(nsw_train[i]))
nsw_train = nsw_train.reshape(60000 * 784, )
stats.describe(nsw_train)
```

```python
def standardize(train):
    mean, std = train.mean(), train.std()
    train = (train - mean) / std
    
    return train
```

```python
s_train = standardize(x_train)
for i in range(4):
    print(stats.describe(s_train[i]))

s_train = s_train.reshape(60000 * 784, )
print(stats.describe(s_train))

```

```python
def standardize_sample_wize(dataset):
    standardized = []
    for s in dataset:
        mean, std = s.mean(), s.std()
        s = (s - mean) / std
        standardized.append(s)
    return np.array(standardized)

ss_train = standardize_sample_wize(x_train)

for i in range(4):
    print(stats.describe(ss_train[i]))

ss_train = ss_train.reshape(60000 * 784, )
print(stats.describe(ss_train))
```

```python
l2_train = kutils.normalize(x_train, axis=0, order=2).reshape(60000 * 784, )
print(stats.describe(l2_train))
```

```python
l1_train = kutils.normalize(x_train, axis=0, order=1).reshape(60000 * 784, )
print(stats.describe(l1_train))
```

```python
l0_train = kutils.normalize(x_train, axis=0, order=0).reshape(60000 * 784, )
print(stats.describe(l0_train))
```

```python
labels, counts = np.unique(x_train, return_counts=True)
print(list(zip(labels, counts)))
```

```python
base_len = len(n_train)
print('Lenght of the initial array: %d' % base_len)
len_without_zeros = len(n_train[n_train > 0.])
print('Lenght without zeros: %d' % len_without_zeros)
print('Percentage of zeros in the dataset: %0.2f' % (len_without_zeros / base_len * 100))
```
